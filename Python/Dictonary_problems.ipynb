Problem 1 — Access Dictionary Value
Problem Statement  Print the value associated with the key 'name'.

Specific Requirements
Use dictionary key access.
Input
student = {
    'name': 'Rahul',
    'age': 22
}
Expected Output
Rahul

```python
input_student = {"name":"Rahul","age":22}
print(f'the name of the student is {input_student["name"]}')
```
