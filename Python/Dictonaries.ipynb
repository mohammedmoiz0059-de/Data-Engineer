### Dictonaries :
    An Unordered collection which is mutable and cannot contain duplicate Keys.
    empty_dict = {}
    holds the value in key:value pairs
    Value can be number,string,a list or even dictionary
    Ex : alien0 = {'color':'green'}


```python
### Accessing a dictionary
alien0 = {'color':'green','points':5}
print(alien0['color'])
print(alien0['points'])
print(f'you just received a alien of color {alien0["color"]} and points {alien0["points"]}')
```

#### Adding a new Key-value pairs
    Dictonaries are mutable we can add/modify/delete.
    To add a new key-value, we give dictonary name followed by [key] = value

```python
alien0 = {'color':'green','points':5}
alien0['lat_loc'] = 0
alien0['lon_loc'] = 15
print(alien0)
```

```python
#Modyfying a dictionary

alien0['color'] = 'red'
print(alien0)
```

```python
#Removing a key-value pair

del alien0['color']
print(alien0)
```

### Looping through Dictionary
    looping through all key-value pairs using .items()
    Looping through all keys using .key()
    looping through all values using .values()

```python
### Looping all key-value
favorite_languages = {
'jen': 'python',
'sarah': 'c',
'edward': 'ruby',
'phil': 'python',
}

for key,value in favorite_languages.items():
    print(f'the key is {key} and the value is {value}')
```

```python
### Looping through keys
favorite_languages = {
'jen': 'python',
'sarah': 'c',
'edward': 'ruby',
'phil': 'python',
}
for k in favorite_languages.keys():
    print(k)

### Sorted keys
for k in sorted(favorite_languages.keys()):
    print(k)
```

```python
# Looping using Values

for v in favorite_languages.values():
    print(v)
```

### Nested Dictonaries


```python
### List of Dictonaries

alien_0 = {'color': 'green', 'points': 5}
alien_1 = {'color': 'yellow', 'points': 10}
alien_2 = {'color': 'red', 'points': 15}
aliens = [alien_0, alien_1, alien_2]

for alien in aliens:
    print(alien)

#### using range to create

aliens =[]
sno = 1
for alien in range(10):
    new_alien = {'color': 'green', 'points': 5,'sno' : sno}
    aliens.append(new_alien)
    sno += 1
print(aliens)




```

```python
# A list in a Dictionary
pizza = {
'crust': 'thick',
'toppings': ['mushrooms', 'extra cheese'],
}
print(f'pizza with toppings {pizza["toppings"]}')
print(f'pizza with toppings {pizza["toppings"][1]}')

```
