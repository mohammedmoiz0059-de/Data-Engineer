### Accessing Elements :
List is an ORDERED collection of Items. Any Element can be accessed by the position or Index.
To access the element write name of list followed by index.

Positive index starts from first 0 to end
Negative index starts from Last  as  -1 to start


```python
print(bicycles[0])
print(bicycles[1])
print(bicycles[-1])
```

```python
#Accessing and Using
print(f'This is my first cycle {(bicycles[0]).upper()}')
```

### Mutable
    List is Mutable collection of items. We can add,remove,modify elements of list.

### Modifying :
    The syntax to modify is similar to accessing, follower by the value.



```python
# Modifying the list
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)
motorcycles[0] = 'ducati'
print(motorcycles)
```

 ### Adding Element :
    Append : Adding elements at the end of the List. list.append(value)
    Insert : Adding element at specified position. list.insert(index,value)

```python
print(motorcycles)
motorcycles.append('honda')
print(motorcycles)
motorcycles.insert(0,'Royal')
print(motorcycles)
```

### Removing Element :
    del : delete an element using index position. del list[index]
    pop() :
        method to delete the last element if no index given. list.pop()
        can remove element at specified position. list.pop(index)
    Remove : To remove an element based on the value. works only first occurence,needs a loop if there are multiple values. list.remove('value')


```python
motorcycles = ['honda', 'yamaha', 'suzuki']
del motorcycles[0]
print(motorcycles)
motorcycles.pop()
print(motorcycles)
motorcycles = ['honda', 'yamaha', 'suzuki']
motorcycles.pop(0)
print(motorcycles)
motorcycles = ['honda', 'yamaha', 'suzuki']
motorcycles.remove('yamaha')
print(motorcycles)
```

### Organizing a list

    Permanent sorting using sort() method. list.sort()
    Temporary sorting using sorted Function. sorted(list)

```python
cars = ['bmw','audi','toyota','suzuki']
print(cars)
cars.sort()
print(cars)

cars = ['bmw','audi','toyota','suzuki']
print(f'The temporary sorted list of cars {sorted(cars)}')

```

### Slicing a List
    Similar to access but takes start:end. list[start:end]


```python
cars = ['bmw','audi','toyota','suzuki']
print(cars[0:2])
```

Reverse a List : To reverse the elements of list. list.reverse()
Length of list : len(list)

```python
cars = ['bmw','audi','toyota','suzuki']
cars.reverse()
print(cars)
print(len(cars))
```

### Looping in a list

```python
cars = ['bmw','audi','toyota','suzuki']
for car in cars :
    if car == 'bmw':
        print(f'{car} is out of stock')
    else:
        print(f'{car} is in stock')
```

### Making a numerical list
    Range : range(start,end,step)

```python
for value in range(1,10):
    print(value)

for value in range(1,10,2):
    print(value)

# TO create a list of numbers

values = list(range(1,10))
print(values)
```

### List Comprehensions or List Comp :
    A concise way of creating list using loops
        [expression for item in list]
        [expression for item in iterable if condition]
        [true_value if condition else false_value for item in iterable]

```python
for value in range(0,10):
    print(value)
```

```python
cars = ['bmw','audi','toyota','suzuki']
for car in cars :
    if car == 'bmw':
        print(f'{car} is out of stock')

[f'out of stock for {car}' for car in cars if car == 'bmw']
```

```python
cars = ['bmw','audi','toyota','suzuki']
for car in cars :
    if car == 'bmw':
        print(f'{car} is out of stock')
    else:
        print(f'{car} is in stock')

[f'out of stock {car}' if car == 'bmw' else f'{car} in stock' for car in cars]
```

### COPYING A LIST
    1. when a list1 = list 2 is done, a refernce copy is done. both the list will be refer to same address. #Deep Copy
    2. when a list is copied using slicing, a new a list is created with new address. #Shallow Copy

```python
cars = ['bmw','audi','toyota','suzuki']
new_cars = cars
print(f'old cars {cars}')
print(f'new cars {new_cars}')
cars.append('Jaguar')
print(f'old cars {cars}')
print(f'new cars {new_cars}')
print(id(cars))
print(id(new_cars))

```

```python
##########################
cars = ['bmw','audi','toyota','suzuki']
new_cars = cars[:]
print(f'old cars {cars}')
print(f'new cars {new_cars}')
cars.append('Jaguar')
print(f'old cars {cars}')
print(f'new cars {new_cars}')
print(id(cars))
print(id(new_cars))
```
