Problem 1 — Reverse a List   
Problem Statement Reverse a list without using reverse().   
Input   
foods = ['pizza', 'burger', 'pasta']   
Expected Output   
['pasta', 'burger', 'pizza']   

```python
#======Using Insert Slower as it has to shift index==========
my_foods = ['pizza', 'falafel', 'carrot cake']
print(my_foods)
reversed = []
for food in my_foods:
    reversed.insert(0,food)
print(reversed)
#=====Using SLicing (Faster as reads in reverse order)=========
reversed_foods = []
for food in my_foods[::-1]:
    reversed_foods.append(food)
print(reversed_foods)
```

Problem 2 — Sum of Elements
Problem Statement

Write a program to find the sum of all numbers in a list without using the built-in sum() function.

Input
numbers = [10, 20, 30, 40]
Expected Output
100

```python
numbers = [10, 20, 30, 40]
sum = 0
for number in numbers:
    sum += number
print(sum)
```

Problem 3 — Find Maximum Number   
Problem Statement   
Write a program to find the largest number in a list without using max().   
Input   
numbers = [12, 45, 7, 89, 23]   
Expected Output   
89  

```python
number = [12,45,7,89,23]
number.sort()
print(number[-1])
#====================
largest = 0

for num in number:
    if num > largest:
        largest =num
print(largest)
```
