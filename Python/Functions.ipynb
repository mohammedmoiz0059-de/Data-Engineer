#### Function
    There are two parts of functions.
        Function definition :  def <name> :
                                    <body of the function>
        Function calling : name()

```python
def greet_user():
    print("hello")

greet_user()
```

#### Parameter and arguments
       when we defined a function definition with a variable -- those are called as parameters
                Parameter
                Global Variable : Created Outside the function, canbe accessed outside
                Local Variable : Created inside the function, cannot be accessed outside
       when we call a function with variable - arguments



```python
f = 2 ## Global Variable
def multiply(x):
    y = x * f
    print(y)

multiply(2)
```

#### Passing Arguments
    1.Positional Arguments : When you call a function, Python must match each argument in the function call with a parameter in the function definition.
        Order Matters in Positional Arguments
        Num of args must match with parameters
    2. Keyword Arguments : is a name-value pair that you pass to a function.

```python
def greeting(name,petname):
    print(f'Hello my name is {name} and my pet name is {petname}')
greeting('maria','puppy')
### Order Matters in Positional Arguments
greeting('puppy','maria')
# Num of args must match with parameters
greeting('maria')

#keyword Arguments
greeting(name='maria',petname='puppy')
```

#### Default Values
    we can define a function with default value

```python
def greeting(name,petname,animal='dog'):
    print(f'Hello my name is {name} and my pet name is {petname} and it is {animal}')
greeting('maria','puppy')
```

#### Return a value
    process some data and then return a value or set of values. The value the function returns is called a return value

```python
# Simple value

def get_formatted_value(first_name,last_name):
    full_name = first_name + ' ' + last_name
    return full_name

name = get_formatted_value('james','bond')
```

```python
### Optional args
def get_formatted_value(first_name,last_name,middle_name=''):
    full_name = first_name + ' ' + middle_name + ' '+ last_name
    return full_name

name = get_formatted_value(first_name='james',last_name='bond')
print(name)
name = get_formatted_value(first_name='james',last_name='bond',middle_name='jim')
print(name)
```

```python
# Passing a list of name

def greet(names):
    for name in names:
        print(f'Hello {name}')

usernames = ['hannah', 'ty', 'margot']
greet(usernames)
```

#### Unknown number of args
    1.Arbitary positional arguments  *args
    2.Arbitary keyword args  **kwargs

```python
def make_pizza(*toppings):
    print(toppings)
make_pizza('pepperoni')
make_pizza('mushrooms', 'green peppers', 'extra cheese')
```

```python
def build_profile(first,last,**user_info):
    profile = {}
    profile['first'] = first
    profile['last'] = last
    for key,value in user_info.items():
        profile[key] = value
    return profile

user_profile = build_profile('albert','einstein',location='princeton')
print(user_profile)
```
