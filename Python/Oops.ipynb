#### Class
    Blueprint of objects. It is a user defined data that bundles Varibles/attributes/data and Methods/Functions.
    Syntax : class name_of_class():
                attributes
                def <methods>:
    __init__(self) : Python runs automatically whenever a new object is created.
                    The self parameter is required in the method definition, and it must come first before the other parameters

```python
class Dog():
    "A simple dog class"

    def __init__(self,name,age):
        "Initialize name and age attributes"
        self.name = name
        self.age = age

    def sit(self):
        "Simulating sitting"
        print(f'{self.name} of {self.age} is sitting')

    def rolling(self):
        "rolling sitting"
        print(f'{self.name} of {self.age} is rolling')


my_dog = Dog('willie', 6)
print(f'my dog is {my_dog.name} and my dog is {my_dog.age}')
my_dog.sit()
my_dog.rolling()

my_dog1 = Dog("Puppy",8)
print(f'my dog is {my_dog1.name} and my dog is {my_dog1.age}')
my_dog1.sit()
my_dog1.rolling()
```

```python
class Car():
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0

    def get_descriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model + ' ' + str(self.odometer_reading) + ' reading'
        return long_name.title()

    def read_odometer(self):
        return "reading of odometer " + str(self.odometer_reading)

my_new_car = Car('audi','a4',2020)
print(my_new_car.get_descriptive_name())
print(my_new_car.read_odometer())
```

#### Inheritance :
    Inherit the attributes and methods from original class - Parent Class.
    With Parent Class methods and attributes and its own attributes and methods - child class

```python
class Car():
    """A simple attempt to represent a car."""
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0
    def get_descriptive_name(self):
        long_name = str(self.year) + ' ' + self.make + ' ' + self.model
        return long_name.title()
    def read_odometer(self):
        print("This car has " + str(self.odometer_reading) + " miles on it.")
    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")
    def increment_odometer(self, miles):
        self.odometer_reading += miles
#The name of the parent class must be included in parentheses in the definition of the child class
class ElectricCar(Car):
    """Represent aspects of a car, specific to electric vehicles."""

#The __init__() method at w takes in the information required to make a Car instance.
    def __init__(self, make, model, year):
        """Initialize attributes of the parent class."""
        super().__init__(make, model, year)
'''The super() function at x is a special function that helps Python make
connections between the parent and child class. This line tells Python to
call the __init__() method from ElectricCar’s parent class, which gives an
ElectricCar instance all the attributes of its parent class. The name super
comes from a convention of calling the parent class a superclass and the
child class a subclass.'''


my_tesla = ElectricCar('tesla', 'model s', 2016)
print(my_tesla.get_descriptive_name())
```
